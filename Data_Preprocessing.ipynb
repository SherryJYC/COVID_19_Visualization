```python
import numpy as np
import pandas as pd
from geopy.geocoders import Nominatim
```

```python
area_df = pd.DataFrame(pd.read_csv("./csv/DXYArea.csv"))
area_df.head()
```

## Process data
* area_df contain both province and city level data, convert them to city level 
* set datetime column
* drop Chinese characters

```python
city_df = area_df.drop(columns=['provinceName', 'province_zipCode','cityName', 'city_zipCode',
                                'province_confirmedCount','province_suspectedCount','province_curedCount',
                               'province_deadCount','updateTime'
                               ])
city_df['DateIime'] = area_df['updateTime']
```

```python
city_df.head()
```

## Add geographical information for visualization

```python
# geolocator is easy to time-out, so partition city_names into 15 parts at first
city_names = city_df['cityEnglishName'].unique()
# delete cities which cannot be located by geolocator
city_names = np.delete(city_names,np.where(city_names=='People from other cities'))
city_names = np.delete(city_names,np.where(city_names=='Ninghe District'))
city_names = np.delete(city_names,np.where(city_names=='Shilifeng Prison'))
city_names = np.delete(city_names,np.where(city_names=='Chongqing High-tech Zone'))
city_names = np.delete(city_names,np.where(city_names=='Qianjiang Tujia and Miao Autonomous County'))
city_names = np.delete(city_names,np.where(city_names=='Xinganmeng'))
city_names = np.delete(city_names,np.where(city_names=='Linguolexi'))
city_names = np.delete(city_names,np.where(city_names=='Xinjiang Production and Construction Corps 4th Division'))
city_names = np.delete(city_names,np.where(city_names=='Xinjiang Production and Construction Corps 7th Division'))
city_names = np.delete(city_names,np.where(city_names=='Shihezi, Xinjiang Production and Construction Corps 8th Division'))
city_names = np.delete(city_names,np.where(city_names=='Xinjiang Production and Construction Corps 9th Division'))
city_names = np.delete(city_names,np.where(city_names=='Xinjiang Production and Construction Corps 12th Division'))
city_names = np.delete(city_names,np.where(city_names=='Wujiaqu, Xinjiang Production and Construction Corps 5th Division'))
# partition
city1 = city_names[:25]
city2 = city_names[25:48]
city3 = city_names[48:75]
city4 = city_names[75:100]
city5 = city_names[100:125]
city6 = city_names[125:150]
city7 = city_names[150:175]
city8 = city_names[175:200]
city9 = city_names[200:225]
city10 = city_names[225:250]
city11 = city_names[250:275]
city12 = city_names[275:300]
city13 = city_names[325:350]
city14 = city_names[350:375]
city15 = city_names[375:]
```

```python
# placeholders for two columns
city_df['Latitude'] = city_df['city_confirmedCount']
city_df['Longitude'] = city_df['city_confirmedCount']
```

```python
# add latitude and longitude for cities (city1 - city15)
for city in city15:
    geolocator = Nominatim(country_bias="China")
    print(city)
    location = geolocator.geocode(city)
    city_df['Latitude'][city_df['cityEnglishName'] == city] = location.latitude
    city_df['Longitude'][city_df['cityEnglishName'] == city] = location.longitude

print("finish this partition")
```
