```python
import pandas as pd
```

```python
data = pd.read_csv('owid-covid-data.csv')
data
```

```python
data_p = data.pivot(index='location',columns='date',values='total_cases')
```

```python
data_p.to_csv('world_pivot.csv')
```

```python
country = pd.read_csv('country.csv')
country
```

```python
data_p = pd.read_csv('world_pivot.csv')
data_p
```

```python
world_join = pd.merge(data_p,country,how='left',left_on='location',right_on='Country Name',right_index=False)
world_join
```

```python
world_join.to_csv('world_join.csv',index=False)
```
