```python
# cure
cure = covid.pivot(index='cityEnglishName',columns='Date',values='city_curedCount')

# fillna with 0/inf
cure0 = np.array(cure.fillna(0))
cureinf = np.array(cure.fillna(float('inf')))

# find max, min index
cure_maxid = np.argmax(cure0,axis=1)
cure_minid = np.argmin(cureinf,axis=1)

# populate na
for i in range(cure0.shape[0]):
    for j in range(cure0.shape[1]):
        if cure0[i,j] == 0 and j <= cure_minid[i]:
            cure0[i,j] = cure0[i,cure_minid[i]]
        elif cure0[i,j] == 0 and j >= cure_maxid[i]: 
            cure0[i,j] = cure0[i,cure_maxid[i]]
        elif cure0[i,j] == 0:
            cure0[i,j] = cure0[i,j-1]

# add city column
cure_processed = pd.DataFrame(cure0)
cure_processed.columns = cure.columns
cure_processed['city'] = cure.index
```

```python
cure_processed_latlon = pd.merge(cure_processed,covid,how='left',left_on='city',right_on='cityEnglishName',left_index=False,right_index=False)
cure_processed_final = cure_processed_latlon.groupby('city').mean().reset_index()
cure_processed_final = cure_processed_final.drop(columns=['city_confirmedCount', 'city_curedCount', 'city_deadCount','city_suspectedCount'])
```

```python
cure_processed_final.to_csv('DXYcity0507_cured.csv')
```
